```python
import os
import PIL.Image as Image
import matplotlib.pyplot as plt
from math import sqrt
path = 'datasets/fire_and_smoke_detect/Fog/train/images'

file_list = os.listdir(path)
file_list= [os.path.join(path, file) for file in file_list]
size_list = []
# get size,hxw
for file in file_list:
    img = Image.open(file)
    size = img.size
    size_list.append(size)
sqrt_size = [sqrt(size[0]*size[1]) for size in size_list]
w = [size[0] for size in size_list]
h = [size[1] for size in size_list]
print(size_list)
# get histogram

plt.hist(sqrt_size, bins=100)
plt.show()
```
